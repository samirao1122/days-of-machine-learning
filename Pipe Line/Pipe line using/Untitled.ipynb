```python
dataset=dataset.drop(columns=["Name","Ticket","Cabin"])
```

```python
x=dataset.drop(columns=["Survived"])
y=dataset["Survived"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=32)
x_train.head(2)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,MinMaxScaler,StandardScaler
```

```python
t1=ColumnTransformer(transformers=[
    ("t1",SimpleImputer(strategy="mean"),[3]),
    ("t3",SimpleImputer(strategy="most_frequent"),[7])
],remainder="passthrough")
t1
```

```python
dataset["Embarked"].value_counts()
```

```python
t2=ColumnTransformer(transformers=[
    ("t2",OneHotEncoder(sparse_output=False,handle_unknown="ignore"),[7])
],remainder="passthrough")
t2
```

```python
t3=ColumnTransformer(transformers=[
    ("t5",StandardScaler(),slice(0,10))
])
t3
```

```python
from  sklearn.pipeline import Pipeline

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

```

```python
pipe=Pipeline(steps=[
    ("t1",t1),
    ("t2",t2),
    ("t3",t3),
#     ("t4",RandomForestClassifier(n_estimators=150,class_weight={0: 1.25, 1: 2}))
    ("t5",LogisticRegression(class_weight={0: 1.25, 1: 2}))
 ])
pipe
```

```python
pipe.fit(x_train,y_train)
```

```python
d=pipe.predict(x_test)
d
```

```python
print("Train Accuracy:", pipe.score(x_train, y_train))
print("Test Accuracy:", pipe.score(x_test,y_test))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# Define the pipeline with a deeper tree and class weighting
pipe = Pipeline(steps=[
    ("t1", t1),
    ("t2", t2),
    ("t3", t3),
    ("t4", DecisionTreeClassifier(class_weight='balanced'))
])

# Optionally, perform hyperparameter tuning
param_grid = {
    't4__max_depth': [None, 5, 10, 20],
    't4__min_samples_split': [2, 5, 10],
    't4__min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(pipe, param_grid, cv=5, scoring='f1')
grid_search.fit(x_train, y_train)

# Evaluate the best model
print("Best Parameters:", grid_search.best_params_)
print("Train Accuracy:", grid_search.best_estimator_.score(x_train, y_train))
print("Test Accuracy:", grid_search.best_estimator_.score(x_test, y_test))
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, roc_auc_score

# Define the pipeline with SMOTE and class weighting
smote = SMOTE(random_state=42)

pipe = ImbPipeline(steps=[
    ("t1", t1),
    ("t2", t2),
    ("t3", t3),
    ("smote", smote),  # Apply SMOTE to balance the classes
    ("t4", DecisionTreeClassifier(class_weight='balanced'))
])

# Fit the pipeline
pipe.fit(x_train, y_train)

# Evaluate the model
y_pred = pipe.predict(x_test)
print(classification_report(y_test, y_pred))
print("AUC-ROC:", roc_auc_score(y_test, y_pred))
```
