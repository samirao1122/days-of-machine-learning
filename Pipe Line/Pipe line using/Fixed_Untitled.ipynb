```python
import pandas as pd
```

```python
dataset=pd.read_csv(r"C:\Users\wwrao\Downloads\train (1).csv")
```

```python
dataset.head(2)
```

```python
dataset=dataset.drop(columns=["Name","Ticket","Cabin"])
```

```python
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
x=dataset.drop(columns=["Survived"])
y=dataset["Survived"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
x_train.head(2)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,MinMaxScaler
```

```python
t1=ColumnTransformer(transformers=[
    ("t1",SimpleImputer(strategy="mean"),[3]),
    ("t3",SimpleImputer(strategy="most_frequent"),[7])
],remainder="passthrough")
t1
```

```python
dataset["Embarked"].value_counts()
```

```python
t2=ColumnTransformer(transformers=[
    ("t2",OneHotEncoder(sparse_output=False, drop="first",handle_unknown="ignore"),[7])
],remainder="passthrough")
t2
```

```python
t3=ColumnTransformer(transformers=[
    ("t5",MinMaxScaler(),slice(0,9))
])
t3
```

```python
from  sklearn.pipeline import Pipeline

from sklearn.tree import DecisionTreeClassifier
```

```python
pipe=Pipeline(steps=[
    ("t1",t1),
    ("t2",t2),
    ("t3",t3),
    ("t4",DecisionTreeClassifier())
])
pipe
```

```python
pipe.fit(x_train,y_train)
```

```python
pipe.predict(x_test)
```
