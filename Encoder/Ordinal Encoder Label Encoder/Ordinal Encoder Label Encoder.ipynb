<h3>Ordinal Encoder</h3>

```python
import pandas as pd 
```

```python
data=pd.read_csv(r"C:\Users\wwrao\Downloads\cleaned_file.csv")
data.head()
```

```python
dataset=data[["Gender", "Married", "Education", "Self_Employed", "Loan_Status"]]
dataset.head()
```

```python
x=dataset.iloc[:,:4]
y=dataset.iloc[:,-1]
x
```

```python

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
x_train.shape
```

```python
from sklearn.preprocessing import OrdinalEncoder

oe=OrdinalEncoder(categories=[["Female","Male"],["No","Yes"],["Graduate","Not Graduate"],["No","Yes"]])
```

```python
oe.fit(x_train)
```

```python
x_train_Scaled=oe.transform(x_train)
x_test_Scaled=oe.transform(x_test)
```

<h3>Lablel Encoder</h3>

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l=LabelEncoder()
```

```python
l.fit(y_train)
```

```python
y_train_Sclaed=l.transform(y_train)
y_test_Sclaed=l.transform(y_test)
```

```python
x_train.columns
```

```python

x_train_df = pd.DataFrame(x_train_Scaled,columns=x_train.columns)
x_test_df = pd.DataFrame(x_test_Scaled,columns=x_train.columns)
```

```python
y_train_df = pd.DataFrame(y_train_Sclaed,columns=["Loan_Status"])
y_test_df = pd.DataFrame(y_test_Sclaed,columns=["Loan_Status"])
```

```python

train_data = pd.concat([x_train_df, y_train_df], axis=1)
test_data = pd.concat([x_test_df, y_test_df], axis=1)
```

```python
dataset = pd.concat([train_data, test_data], axis=0)
```
