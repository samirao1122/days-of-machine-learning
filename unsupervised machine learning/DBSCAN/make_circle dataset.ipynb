```python
from sklearn.datasets import make_circles
```

```python
x,_=make_circles(n_samples=500,noise=0.8)
```

```python
import pandas as pd
dataset=pd.DataFrame(x,columns=["a","b"])
```

```python
dataset.head(2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(dataset["a"],dataset["b"])
```

```python
from sklearn.cluster import DBSCAN
```

```python
d=DBSCAN(eps=0.3,min_samples=6)
```

```python
y_label=d.fit_predict(dataset)
```

```python
y_label
```

```python
plt.scatter(dataset["a"],dataset["b"],c=y_label)
```
