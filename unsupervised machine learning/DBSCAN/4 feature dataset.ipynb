```python
import pandas as pd
```

```python
dataset=pd.read_csv(r"C:\Users\wwrao\Downloads\kmeans_4D_dataset.csv")
```

```python
dataset.head()
```

```python
import matplotlib.pyplot as plt 
```

```python
from  sklearn.decomposition import PCA
```

```python
p=PCA(n_components=2
     )
```

```python
data=p.fit_transform(dataset.iloc[:,:-1])
```

```python
df=pd.DataFrame(data,columns=["a","b"])
```

```python
df.head(2)
```

```python
plt.scatter(df.a,df.b)
```

```python
from sklearn.cluster import DBSCAN
```

```python
d=DBSCAN(eps=1.3,min_samples=10)
```

```python
y_label=d.fit_predict(df)
```

```python
plt.scatter(df.a,df.b,c=y_label)
```
