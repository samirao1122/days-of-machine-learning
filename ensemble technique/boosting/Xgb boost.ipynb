```python
import pandas as pd 
```

```python
dataset=pd.read_csv(r"C:\Users\wwrao\Downloads\most_complex_regression_dataset.csv")
dataset.head(2)
```

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

```python
x=dataset.drop(columns=["Target"])
x.head()
```

```python
y=dataset["Target"]
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
import seaborn as sns 
sns.pairplot(dataset)
```

```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
l=LinearRegression()
lasso=Lasso()
r=Ridge()
sv=SVR()
d=DecisionTreeRegressor()
```

```python
l.fit(x_train,y_train)
```

```python
r.fit(x_train,y_train)
```

```python
lasso.fit(x_train,y_train)
```

```python
d.fit(x_train,y_train)
```

```python
sv.fit(x_train,y_train)
```

**on training**

```python
print("Accuracy score of linear Regression>>>>>>>>")
print(l.score(x_train,y_train)*100)
print("Accuracy score of lasso Regression>>>>>>>>>")
print(lasso.score(x_train,y_train)*100)
print("Accuracy score of Ridge  Regression>>>>>>>>")
print(r.score(x_train,y_train)*100)
print("Accuracy score of svR  Regression>>>>>>>>")
print(sv.score(x_train,y_train)*100)

print("Accuracy score of DecisonTreeRegression>>>>>>>>")
print(d.score(x_train,y_train)*100)
```

```python
y_predict=l.predict(x_test)
y_predict
```

```python
y_test
```

```python
from sklearn.metrics  import r2_score
r2_score(y_test,y_predict)
```

```python
sns.displot(y_test-y_predict,kind="kde")
```
